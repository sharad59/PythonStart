```python
#MERGEER of companies and rounds2

import pandas as pd
companies = pd.read_csv("companies.txt", sep="\t", encoding = "ISO-8859-1")
companies_new = companies.rename(columns={'permalink': 'company_permalink'})
companies_new["company_permalink"] = companies_new["company_permalink"].str.lower()
rounds2["company_permalink"] = rounds2["company_permalink"].str.lower()
master_frame = pd.merge(rounds2, companies_new, how='inner', on='company_permalink')

x=master_frame['company_permalink'].nunique()
print(x)
master_frame.head()
```

```python
#Checkpoint 3: Country Analysis
#Spark Funds wants to see the top nine countries which have received the highest total funding
import pandas as pd

country = master_frame.pivot_table(values = 'raised_amount_usd', index = 'country_code', aggfunc = 'sum')

unsorted = pd.DataFrame(data = country)

top9= pd.DataFrame(data = unsorted.nlargest(9, 'raised_amount_usd'))

top9.head()

```
